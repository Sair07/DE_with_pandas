# Getting Data from multiple sheet

* By defualt read_excel loads only first sheet
* Sheet are zero index
* accept names & index - start from 0
* can be passed as lists
* Becareful while passing n sheet with one read_excel - as it applies all paremeter to all n sheets
* if you want different with different parameters or attriubtes call different read_excel

```python
#getting second sheet
#sheet name - 2017 or as index -1 (as its second sheet) or sheet_name = '2017'
data_sheet2 = pd.read_excel('/Users/apple/Documents/Python/datasets/fcc-new-coder-survey.xlsx', sheet_name = 1,skiprows=2,
                           usecols = "A:CT")
data_sheet2.head(3)
```

```python
data_sheet2.shape
```

```python
#Loading all sheet
all_data = pd.read_excel('/Users/apple/Documents/Python/datasets/fcc-new-coder-survey.xlsx', sheet_name = None, skiprows =2)
type(all_data)
```

* Loading all sheet, return you a dict and not a DataFrame

```python
all_data['2016'].head()
```

```python
#our excel both sheet have same columns .. We can merge the data in one dataFrame as below

Full_Data = pd.DataFrame()
Full_Data = pd.concat([all_data['2016'],all_data['2017']])
```

```python
Full_Data.shape
```

```python
#looping through multiple sheet and adding it to one DataFram 
#use concat for 2 sheet.. 

df1 = pd.DataFrame()
# remember the value is the dataframe 
for sn, frame in all_data.items():
    #adding year as column to get which sheet it came from
    frame['Year'] = sn 

    df1 = pd.concat([df1, frame])
df1.shape
```

# DateTime formats 

* use pd.to_datetime -- to convert
* you can pass the formats - in what format you want your data be displayed in
* https://www.bairesdev.com/tools/strftime/ has a list of all DT formats

```python
df1[['Part1EndTime','Part1StartTime', 'Part2EndTime', 'Part2StartTime']].dtypes
```

```python
format_string = "%Y-%m-%d %H:%M:%S"
df1['Part1EndTime'] = pd.to_datetime(df1['Part1EndTime'],format = format_string)
```
