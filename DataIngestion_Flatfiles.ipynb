```python
#suppose you want to change zipcode to str type
# use dtype - pass as dict {"column_name": type}

cols = ['STATEFIPS', 'STATE', 'zipcode', 'agi_stub', 'N1' ] # we can also use [1,2,3,4,5]
data = pd.read_csv('/Users/apple/Documents/Python/datasets/vt_tax_data_2016.tsv',usecols = cols, dtype = {"zipcode": str})
data.info()
```

```python
data_types = {"agi_stub": "category",
              "zipcode": str}

# Load csv using dtype to set correct data types
data = pd.read_csv("/Users/apple/Documents/Python/datasets/vt_tax_data_2016.tsv", dtype = data_types)

# Print data types of resulting frame
print(data.dtypes.head())
```

```python
#pandas interprets missing value as NA or Null 

# na_values accepts single val, list, dict 
# anything 0 empty just a invalid charater - will be converted to NA

data.sort_values('zipcode')
```

```python
#handling 0's before loading data 
# where ever its 0 itll be now null
cols = ['STATEFIPS', 'STATE', 'zipcode', 'agi_stub', 'N1' ] # we can also use [1,2,3,4,5]
data = pd.read_csv('/Users/apple/Documents/Python/datasets/vt_tax_data_2016.tsv',usecols = cols, na_values = {"zipcode": 0})
data[data['zipcode'].isna()]
```
