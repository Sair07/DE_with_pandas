# Different JSON orient

| Orient    | Typical usage                      |
| --------- | ---------------------------------- |
| `records` | APIs / REST                        |
| `split`   | efficient DataFrame reconstruction |
| `columns` | default pandas format              |
| `table`   | standardized exchange              |
| `values`  | matrix-like data                   |
